Q7: Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name.

```python
sql = """SELECT DISTINCT(Members.firstname || ' ' || Members.surname) as name, Facilities.name as court
FROM Bookings
INNER JOIN Facilities
ON Bookings.facid = Facilities.facid
INNER JOIN Members
ON Bookings.memid = Members.memid
Where Facilities.facid = 0 OR Facilities.facid = 1
ORDER BY name
"""
q7 = pd.read_sql_query(sql, connection)
q7
```

Q8: Produce a list of bookings on the day of 2012-09-14 which
will cost the member (or guest) more than $30. Remember that guests have
different costs to members (the listed costs are per half-hour 'slot'), and
the guest user's ID is always 0. Include in your output the name of the
facility, the name of the member formatted as a single column, and the cost.
Order by descending cost, and do not use any subqueries.

```python
sql = """ SELECT Facilities.name as facility, (Members.firstname || ' ' || Members.surname) as fullname,
CASE WHEN Bookings.memid = 0 THEN Bookings.slots * Facilities.guestcost
ELSE Bookings.slots * Facilities.membercost
END AS cost
FROM Bookings
LEFT JOIN Facilities
ON Bookings.facid = Facilities.facid
LEFT JOIN Members
ON Bookings.memid = Members.memid
WHERE Bookings.starttime LIKE '2012-09-14%' and cost>30
Order BY cost DESC
"""
q8 = pd.read_sql_query(sql, connection)
q8
```

Q9: This time, produce the same result as in Q8, but using a subquery.

```python
sql = """ SELECT facility, fullname, cost, starttime
FROM (Select f.name AS facility, 
(m.firstname || ' ' || m.surname) as fullname, 
b.starttime AS starttime,
CASE WHEN b.memid = 0 THEN b.slots * f.guestcost
ELSE b.slots * f.membercost
END AS cost
FROM Bookings AS b
Inner Join Facilities AS f
ON b.facid = f.facid
Inner Join Members AS m
ON b.memid = m.memid
WHERE cost > 30 and starttime LIKE '2012-09-14%'
ORDER BY cost DESC) as subquery

"""
q9 = pd.read_sql_query(sql, connection)
q9
```

Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
sql = """
SELECT Facilities.name as facility,
(SELECT SUM(CASE WHEN Bookings.memid = 0 THEN Bookings.slots * Facilities.guestcost
ELSE Bookings.slots * Facilities.membercost
END)) AS revenue
FROM Bookings
LEFT JOIN Facilities
ON Bookings.facid = Facilities.facid
LEFT JOIN Members
ON Bookings.memid = Members.memid
GROUP BY Facilities.name
HAVING revenue < 1000
ORDER BY revenue
"""
q10 = pd.read_sql_query(sql, connection)
q10
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
sql = """
SELECT m.surname AS Member_surname, m.firstname AS Member_Firstname, recommended.surname AS Recommended_surname, recommended.firstname AS Recommended_firstname
FROM Members AS m 
INNER JOIN Members AS recommended
ON m.memid = recommended.recommendedby
ORDER BY Member_surname, Member_firstname
"""
q11 = pd.read_sql_query(sql, connection)
q11
```

Q12: Find the facilities with their usage by member, but not guests

```python
sql = """
SELECT (m.firstname || ' ' || m.surname) as fullname, f.name AS facility, sum(b.slots) AS usage
FROM Bookings AS b
LEFT JOIN Facilities AS f
ON b.facid = f.facid
LEFT JOIN Members AS m
ON b.memid = m.memid
WHERE b.memid <> 0
GROUP by fullname
ORDER BY usage, fullname
"""
q12 = pd.read_sql_query(sql, connection)
q12
```

Q13: Find the facilities usage by month, but not guests
Used Substring() to extract year and month

```python
sql = """
SELECT SUBSTRING(b.starttime, 1, 7) AS months, SUM(b.slots) as usage, f.name AS facility
FROM Bookings AS b
LEFT JOIN Facilities AS f
ON b.facid = f.facid
LEFT JOIN Members AS m
ON b.memid = m.memid
GROUP by facility, months
ORDER BY months;
"""
q13 = pd.read_sql_query(sql, connection)
q13
```
